```python
import re

# Задание 1 — Проверка ИНН
def validate_inn():
    while True:
        inn = input("Введите ИНН (12 цифр): ").strip()
        if re.fullmatch(r'\d{12}', inn):
            print(f"✅ ИНН принят: {inn}")
            return inn
        else:
            print("❌ ИНН должен состоять ровно из 12 цифр.")

# Задание 2 — Валидация цены
def validate_price():
    while True:
        try:
            price = float(input("Введите цену товара: "))
            if 0 < price <= 1_000_000:
                print(f"✅ Цена принята: {price}")
                return price
            else:
                print("❌ Цена должна быть положительной и не превышать 1 000 000.")
        except ValueError:
            print("❌ Введите числовое значение.")

# Задание 3 — Валидация пароля
def validate_password():
    while True:
        password = input("Введите пароль (8-20 символов, заглавная буква и цифра): ")
        if not (8 <= len(password) <= 20):
            print("❌ Длина пароля должна быть от 8 до 20 символов.")
        elif not re.search(r'[A-Z]', password):
            print("❌ Пароль должен содержать хотя бы одну заглавную букву.")
        elif not re.search(r'\d', password):
            print("❌ Пароль должен содержать хотя бы одну цифру.")
        else:
            print(f"✅ Пароль принят.")
            return password

# Тест
validate_inn()
validate_price()
validate_password()
```
